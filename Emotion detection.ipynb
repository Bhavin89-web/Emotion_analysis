```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

img1=cv2.imread('images1.jpg')
plt.imshow(img1)
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))

#predicting emotions-happy
predictions=DeepFace.analyze(img1)

predictions

#detecting faces
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 
                                     'haarcascade_frontalface_default.xml')

gray_img=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)

faces=face_cascade.detectMultiScale(gray_img,scaleFactor=1.12,minNeighbors=5)

for x,y,w,h in faces:
    img=cv2.rectangle(gray_img,(x,y),(x+w,y+h),(255,255,0),4)
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img1,predictions['dominant_emotion'],(0,50),font,1,(0,0,255),2,cv2.LINE_4);

plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
```

```python
import cv2  
from deepface import DeepFace
  
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 
                                     'haarcascade_frontalface_default.xml')
  
# capture frames from a camera 
cap = cv2.VideoCapture(0) 

#if not cap.isOpened():
   # cap=cv2.VideoCapture(0)
#if not cap.isOpened():
   # raise IOError('Cannot open webcam')
  
# loop runs if capturing has been initialized. 
while True:  
  
    # reads frames from a camera 
    ret,frame = cap.read()  
    result=DeepFace.analyze(frame,actions=['emotion'],enforce_detection=False)
  
    # convert to gray scale of each frames 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
  
    # Detects faces of different sizes in the input image 
    faces = face_cascade.detectMultiScale(gray, 1.1, 5) 
  
    for (x,y,w,h) in faces: 
        # To draw a rectangle in a face  
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),6)  
    
    font=cv2.FONT_HERSHEY_SIMPLEX
    
    cv2.putText(frame,result['dominant_emotion'],(50,50),font,3,(0,0,255),2,cv2.LINE_4);
    
    cv2.imshow('Demo video',frame)
    # Display an image in a window 
    
    if cv2.waitKey(2) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
